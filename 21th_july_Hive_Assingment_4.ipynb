```python
# 1.  Write a Python program that uses the HiveQL language to create a table named "Employees" with columns for "id," "name,
# " and "salary."


from pyhive import hive

# Connect to Hive server
conn = hive.Connection(host='localhost', port=10000, username='your_username')

# Create table query
create_table_query = """
    CREATE TABLE Employees (
        id INT,
        name STRING,
        salary FLOAT
    )
"""

# Execute the query
with conn.cursor() as cursor:
    cursor.execute(create_table_query)

# Close the connection
conn.close()
```

```python
# 2.  Create a Python program that retrieves records from a Hive table named "Customers" where the age is greater than 30.


from pyhive import hive

# Connect to Hive server
conn = hive.Connection(host='localhost', port=10000, username='your_username')

# Query to retrieve records
retrieve_records_query = """
    SELECT *
    FROM Customers
    WHERE age > 30
"""

# Execute the query
with conn.cursor() as cursor:
    cursor.execute(retrieve_records_query)
    records = cursor.fetchall()

# Process the retrieved records
for record in records:
    print(record)

# Close the connection
conn.close()
```

```python
# 3.  Write a Python script that sorts records in descending order based on the "timestamp" column in a Hive table named
#"Logs."


from pyhive import hive

# Connect to Hive server
conn = hive.Connection(host='localhost', port=10000, username='your_username')

# Query to sort records
sort_records_query = """
    SELECT *
    FROM Logs
    ORDER BY timestamp DESC
"""

# Execute the query
with conn.cursor() as cursor:
    cursor.execute(sort_records_query)
    sorted_records = cursor.fetchall()

# Process the sorted records
for record in sorted_records:
    print(record)

# Close the connection
conn.close()
```

```python
# 5.  Write a Python script that calculates the average salary of employees from a Hive table named "Employees".



from pyhive import hive

# Connect to Hive server
conn = hive.Connection(host='localhost', port=10000, username='your_username')

# Query to calculate average salary
calculate_average_salary_query = """
    SELECT AVG(salary) AS average_salary
    FROM Employees
"""

# Execute the query
with conn.cursor() as cursor:
    cursor.execute(calculate_average_salary_query)
    result = cursor.fetchone()

# Process the result
average_salary = result[0]
print("Average Salary:", average_salary)

# Close the connection
conn.close()
```

```python
# 6.  Implement a Python program that uses Hive partitioning to create a partitioned table named "Sales_Data" based on the "
# year" and "month" columns.


from pyhive import hive

# Connect to Hive server
conn = hive.Connection(host='localhost', port=10000, username='your_username')

# Create table query with partitioning
create_table_query = """
    CREATE TABLE Sales_Data (
        id INT,
        product STRING,
        sales INT
    )
    PARTITIONED BY (year INT, month INT)
"""

# Execute the query
with conn.cursor() as cursor:
    cursor.execute(create_table_query)

# Close the connection
conn.close()
```

```python
# 8.  Create a Python program that performs an inner join between two Hive tables, "Orders" and "Customers," based on a common column.


from pyhive import hive

# Connect to Hive server
conn = hive.Connection(host='localhost', port=10000, username='your_username')

# Query to perform inner join
inner_join_query = """
    SELECT *
    FROM Orders
    INNER JOIN Customers ON Orders.customer_id = Customers.id
"""

# Execute the query
with conn.cursor() as cursor:
    cursor.execute(inner_join_query)
    joined_records = cursor.fetchall()

# Process the joined records
for record in joined_records:
    print(record)

# Close the connection
conn.close()
```
